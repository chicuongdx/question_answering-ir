```python
DRIVE_PATH = '/content/gdrive/MyDrive/question_answering_data/'
GROUND_TRUTH = 'zac2022_test_ground_truth.json'
```

```python
import pandas as pd
import json
from tqdm.auto import tqdm
```

```python
def calculate_accuracy(ground_truth_file, test_file):

  with open(DRIVE_PATH + ground_truth_file) as f:
    ground_truth = json.load(f)['data']
  ground_truth = pd.DataFrame(ground_truth)

  with open(DRIVE_PATH + test_file) as f:
    test = json.load(f)
  test = pd.DataFrame(test)

  num_rows = len(ground_truth)
  num_correct = 0
  for (idx_truth, row_truth), (idx_test, row_test) in zip(ground_truth.iterrows(), test.iterrows()):

    if row_truth['answer'] == row_test['answer']:
      num_correct += 1

  return num_correct/num_rows
```

```python
calculate_accuracy(GROUND_TRUTH, 'test_from_model/test_tokenizer.json')
```
