```python
DRIVE_PATH = '/content/gdrive/MyDrive/question_answering_data/'
```

```python
import pandas as pd
import json
```

```python
with open(DRIVE_PATH + 'zac2022_train_merged_final.json') as f:
  data = json.load(f)['data']
  data = pd.DataFrame(data)

data
```

```python
all_texts = []
all_questions = []
labels = []
for idx, x in data.iterrows():
  print('Entity', idx)
  all_texts.append(x['text'])
  all_questions.append(x['question'])
  if x['is_long_answer'] or x['category'] == 'FULL_ANNOTATION':
    labels.append(1)
  else:
    labels.append(0)

df = pd.DataFrame({'text': all_texts,'question': all_questions, 'label': labels})
df.text = df.text.apply(lambda x: x.replace("\r","\n"))
df
```

```python
df.to_csv(DRIVE_PATH + "processed/train_ranking.csv",index=False)
```
